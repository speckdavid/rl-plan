```python
import os
import json
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
sb.set_style("darkgrid")
sb.set_context("paper",  font_scale = 1,
               rc={
                   "grid.linewidth": 2.5,
                   'axes.labelsize': 24,
                   "axes.titlesize": 26,
                   "legend.fontsize": 24.0,
                   'lines.linewidth': 2.5,
                   'xtick.labelsize': 22.0,
                   'ytick.labelsize': 22.0,
               })
```

```python
exp_dir = '/mhome/biedenka/rl-plan-tmp/experiments'
for inst in range(21):
    df = pd.read_csv(os.path.join(exp_dir, 'instance%d' % inst, 'scores.txt'), sep='\t', index_col=None)
    plt.step(df.steps, df['mean'] * -1, where='mid', label='learned')
    plt.plot(df.steps.loc[[0, df.shape[0] -1]], [opt_cost_dict[inst], opt_cost_dict[inst]],
             linestyle='--', label='optimal')
    plt.title('instance%d' % inst)
    plt.xlabel('#steps')
    plt.ylabel('Plan-Cost')
    plt.semilogy()
    plt.legend()
    plt.show()
    print('Optimal policy: ')
    with open(os.path.join(exp_dir, 'artifical-benchmarks', 'train', 'instance%d' % inst, 'optimal.json')) as fh:
        print(json.load(fh)[1:])
    print('#'*80)
```
